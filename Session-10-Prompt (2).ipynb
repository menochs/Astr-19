```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# pull 1000 random numbers 
data = np.random.normal (loc=0, scale=1,size=1000)
```

```python
# Histogram the random numbers into 100 bins, and plot the histogram
plt.hist(data, bins=100)
# label your axes
plt.xlabel ("Value")
plt.ylabel ("Frequency")
plt.title("Histogram of Random Numbers from a Normal Distribution")
plt.show()
```

```python
# save figure as a PDF
plt.savefig('distribution_histogram.pdf')
```
