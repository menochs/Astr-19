```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# pull 1000 random numbers distributed uniformly between [0,1]
data = np.random.normal(loc=0, scale=1,size=1000)
```

```python
# Histogram the random numbers into 100 bins, and plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=100, alpha=0.75)
plt.xlabel ("Value")
plt.ylabel ("Frequency")
plt.title("Histogram of Random Numbers from a Normal Distribution")
plt.grid(True)
plt.show()
```
