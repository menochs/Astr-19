```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# pull 1000 random numbers distributed uniformly between [0,1]
random_numbers = np.random.uniform (0, 1, 1000)
```

```python
# Histogram the random numbers into 100 bins, and plot the histogram
plt.hist(random_numbers, bins=100)
plt.xlabel ("Number")
plt.ylabel ("Frequency")
plt.title("Uniform Distribution of 1000 Random Numbers")
```

```python
plt.savefig('random_numbers_histogram.pdf')
```

```python
plt.show()
```
